```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('wyniki.csv')
```

```python
fig, axs = plt.subplots(3, 1, figsize=(7, 21))

for col, ax in zip(('pattern_len', 'text_len', 'dictionary_len'), axs):
    sns.pointplot(
        data=df, x=col, y="time[ns]", hue="function", errorbar=None, ax=ax
    )
```
